## Problem 2a (2 points)

Write a function called *compute_area* that takes two arguments:  the length of a rectangle and the width of a rectangle and returns the area of the rectangle--it should **return** this value, not print it.

```python
# your answer to 2a goes here
def compute_area (length, width):
    area = length * width
    return area
```

## Problem 3 (4 points)

Write a function *is_negative* that takes a number as its only argument and returns the Python Boolean value True if the number is negative, and False otherwise. (The number 0 is neither positive nor negative.)  For full credit **do not use an if statement** in your code.

```python
# Your answer to #3 goes here
def is_negative(num):
    while num<0:
        return True
    else:
        return False
```

## Problem 4 (4 points)

Write a function called *fv* that takes three arguments (*r*, *n*, *pv*) to calculate and return the future value of a lump sump *pv* invested at the periodic rate *r* for *n* periods.  The formula is
$FV = PV(1+r)^n $.

```python
# Your answer to #4 goes here
def fv(r,n,pv):
    total = (pv)* (1+r)**n
    return total
```
