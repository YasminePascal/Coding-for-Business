```python
import ipywidgets as widgets
import requests
import re 
from ipywidgets import AppLayout, Button, Layout
from IPython import display as disp

## show selection

main_out = widgets.Output()
text2_out = widgets.Output()
category_w= widgets.Text(
    placeholder = "Enter the name of a T.V. show", 
    description = "Shows",
    disabled = False
)
search = widgets.Button(description= "Search!")
search.style.button_color = "lightblue"

header_button= widgets.Button(description = "TV MAZE",layout= widgets.Layout(width = "900px"),button_style = "success")
header_button.style.font_style= "italic"
header_button.style.font_weight = "bold"
header_button.style.text_decoration = "underline"
header_button.style.font_size = "25pt"

regex= re.compile("<.*?>")

def remove_html(string):
    clean_text= re.sub(regex,'',string)
    return clean_text

def show_chosen(b):
    with text2_out:
        site = " https://api.tvmaze.com/search/shows"
        apikey= "e4Lnf6_zhI1JPbKR4pduBdj2jE9B58PU"
        category= category_w.value.title()
        parameters= {"q": category, "apikey":apikey}
        r= requests.get(site,parameters)
        show_data = r.json()
        cut = show_data[:5]
        y= 0
        for show in cut:
            y+=1
            print(f'Score:{show["score"]}')
            print(f'Name:{show["show"]["name"]}')
            print(f'Gennre:{show["show"]["genres"]}')
            print(f'Language(s):{show["show"]["language"]}')
            try:
                print(f'Summary:{remove_html(show["show"]["summary"])}')
            except:
                pass 
            print(f'Premiered Date: {show["show"]["premiered"]}')
            print(f'End date: {show["show"]["ended"]}')
            try:
                url2=show["show"]["image"]["medium"]
                display(disp.Image(url2,width= 250))
            except:
                pass
            print('------------------------------------------------')
            print()
        if y==0:
            print("No matches found")
            
        text2_out.clear_output(wait=True)    
        
search.on_click(show_chosen)
category_w.observe(search,names="value")

with main_out:
    display(header_button,category_w,search,text2_out)
    
##celebrity selection

main_out2= widgets.Output()
text_out3= widgets.Output()
people_w= widgets.Text(
    placeholder= "Enter the name of an actor or actress",
    description= "Celebrities",
    disabled = False,
    layout = widgets.Layout(width= "320px")
)

search2 = widgets.Button(description= "Search!")
search2.style.button_color= "yellow"

def chosen_celeb(a):
    with text_out3:
        site2 = "https://api.tvmaze.com/search/people"
        apikey= "e4Lnf6_zhI1JPbKR4pduBdj2jE9B58PU"
        people= people_w.value.title()
        parameter = {"q":people,"apikey":apikey}
        s = requests.get(site2,parameter)
        celeb_data = s.json()
        z = 0
        
        for celeb in celeb_data:
            z+=1
            print(f'Name:{celeb["person"]["name"]}')
            try:
                print(f'Place of birth:{celeb["person"]["country"]["name"]}')
            except:
                pass
                      
            print(f'Date of birth:{celeb["person"]["birthday"]}')
            print(f'Date of death:{celeb["person"]["deathday"]}')
            
            try:
                url=celeb["person"]["image"]["medium"]
                display(disp.Image(url,width= 250))
            except:
                pass
            print("------------------------------------")
            print()
        if z == 0:
            print("No matches found")
        text_out3.clear_output(wait=True)

search2.on_click(chosen_celeb)
people_w.observe(search2,names="value")

with main_out2:
    display(header_button,people_w,search2,text_out3)
    

##filters

main_out3= widgets.Output()
text_out = widgets.Output()

genre_w= widgets.Dropdown(
    options = ["Action","Adult","Adventure","Anime","Children","Comedy","Crime","Drama","Fantasy","Horror","Romance","Supernatural","Thriller","Western"],
    value = "Action",
    description= "Genre:",
    disabled = False
)
language_w= widgets.Dropdown(
    options= ["Arabic","Bulgarian","Chinese","Danish","English","French","German","Hebrew","Italian","Japanese","Korean","Spanish"],
    value = "Arabic",
    description= "Language:",
    disabled = False
)

status_w= widgets.Dropdown(
    options= ["Running","Ended"],
    value = "Running",
    description = "Show Status:",
    disabled= False
)

search3 = widgets.Button(description= "Search!")
search3.style.button_color= "orange"

def chosen_options(c):
    with text_out:
        site3= "https://api.tvmaze.com/shows"
        apikey= "e4Lnf6_zhI1JPbKR4pduBdj2jE9B58PU"
        parameter2 = {"apikey":apikey}
        q = requests.get(site3,parameter2)
        data = q.json()
        genre= genre_w.value.title()
        language= language_w.value.title()
        status= status_w.value.title()
        x=0
        for show in data:
            if genre in show["genres"]:
                if language in show["language"]:
                    if status in show["status"]:
                        x+=1
                        print(f'Name:{show["name"]}')
                        print(f'Genre:{show["genres"]}')
                        print(f'Language:{show["language"]}')
                        print(f'Show type:{show["type"]}')
                        print(f'Show status:{show["status"]}')
                        print(f'Rating:{show["rating"]["average"]}')
                        try:
                            print(f'Summary:{remove_html(show["summary"])}') 
                        except:
                            pass
                        try:
                            url3=show["image"]["medium"]
                            display(disp.Image(url3,width= 250))
                        except:
                            pass
                        
                        print("-------------------------------------")
                        print()
        if x==0:
            print("No matches found")
            
        text_out.clear_output(wait=True) 
        
search3.on_click(chosen_options)
genre_w.observe(search3,names="value")
language_w.observe(search3,names="value")
status_w.observe(search3,names="value")

with main_out3:
    display(header_button,genre_w,language_w,status_w,search3,text_out)
    

##display
children = [main_out,main_out2,main_out3]
tab= widgets.Tab()
tab.children=children
tab.titles= ["Show Selection","People","Filters"]
tab


```
